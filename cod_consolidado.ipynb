```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# 1. CARGA Y LIMPIEZA DE DATOS
# 1.1 Cargamos el dataset
df = pd.read_csv("dataset.txt")
df.isnull().sum()  # Verificamos nulos en el dataset "Total Price"
```

```python
# 1.2 Eliminacion de nulos
df = df.dropna()  # Eliminamos filas con nulos
df.isnull().sum()
```

```python
# 2. OBJETIVO ESPECIFICO 1:
# Analizar el promedio de volúmenes de ventas por cada tienda para identificar 
# las sedes con mayor eficiencia y volumen de movimiento.
```

```python
# 2.1 CALCULO DE METRICAS POR TIENDA
# Agrupamos una sola vez para obtener media de unidades e ingresos totales
# 2.2 ANÁLISIS DESCRIPTIVO (Objetivo 1)
print(f"{' RESUMEN ESTADÍSTICO ':-^50}")
print(df.describe().round(3))
```

```python
# Cálculo de métricas por tienda
## Agrupamos en base "Store ID" y calculamos la "media de unidades vendidas" e "ingresos totales"
df["Ingreso_Total"] = df["Total Price"] * df["Units Sold"]

stats_tiendas = df.groupby("Store ID").agg({
    "Units Sold": "mean",
    "Ingreso_Total": "sum"}).rename(columns={"Units Sold": "Media_Ventas", "Ingreso_Total": "Ingresos_Suma"})
# 2.3 VISUALIZACIÓN DE RESULTADOS
## Configuramos el estilo visual de los gráficos
plt.style.use('seaborn-v0_8-muted')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

## --- Gráfico 1: Top 3 Ventas ---
top_3_ventas = stats_tiendas["Media_Ventas"].nlargest(3)
bars1 = ax1.bar(top_3_ventas.index.astype(str), top_3_ventas.values, color=['#4C72B0', '#55A868', '#C44E52'])
ax1.set_title("Top 3: Promedio de Unidades Vendidas", fontsize=12, fontweight='bold')
ax1.set_xlabel("ID de Tienda")
ax1.set_ylabel("Unidades")
ax1.bar_label(bars1, fmt='%.0f', padding=3)

## --- Gráfico 2: Top 3 Ingresos ---
top_3_ingresos = stats_tiendas["Ingresos_Suma"].nlargest(3)
bars2 = ax2.bar(top_3_ingresos.index.astype(str), top_3_ingresos.values, color=['#8172B3', '#CCB974', '#64B5CD'])
ax2.set_title("Top 3: Ingresos Totales Generados", fontsize=12, fontweight='bold')
ax2.set_xlabel("ID de Tienda")
ax2.set_ylabel("Monto Total ($)")

## Formato de moneda para las etiquetas
ax2.bar_label(bars2, labels=[f'S/ {x:,.0f}' for x in top_3_ingresos], padding=3)

plt.tight_layout()
plt.show()

## Resumen final en consola
print("OBJETIVO ESPECIFICO 2")
print("*"*100+"\n")
```

```python
# 3. OBJETIVO ESPECIFICO 2:
# Determinar la relación entre las variables de precio base y el precio total en la cantidad de productos vendidos
```

```python
# 3.1. CREACIÓN DE VARIABLES LOGARÍTMICAS
# Eliminamos nulos y valores <= 0 para evitar errores en logaritmos
data = df.dropna(subset=["Total Price", "Base Price", "Units Sold"]).copy()
data = data[(data["Total Price"] > 0) & (data["Base Price"] > 0) & (data["Units Sold"] > 0)]
print("✅ Datos cargados y filtrados correctamente.")
```

```python
# 3.2. CREACIÓN DE VARIABLES LOGARÍTMICAS
data["Log_Base_Price"] = np.log(data["Base Price"])
data["Log_Total_Price"] = np.log(data["Total Price"])
data["Log_Units_Sold"] = np.log(data["Units Sold"])
```

```python
# 3.3. FUNCIÓN DE REGRESIÓN
def ejecutar_regresion(y_var, x_vars, titulo):
    print(f"\n{' ' + titulo + ' ':=^80}")
    Y = data[y_var]
    X = sm.add_constant(data[x_vars])  # Usamos x_vars directamente de los argumentos
    modelo = sm.OLS(Y, X).fit()
    print(modelo.summary())
```

```python
# 3.4. Ejecutar Modelo LOG-LOG
ejecutar_regresion('Log_Units_Sold', ['Log_Base_Price', 'Log_Total_Price'], "MODELO LOG-LOG")

# 3.5. Ejecutar Modelo LIN-LIN
ejecutar_regresion('Units Sold', ['Base Price', 'Total Price'], "MODELO LIN-LIN")
```

```python
# 3.6. VISUALIZACIÓN DE HISTOGRAMAS
variables = ["Base Price", "Total Price", "Units Sold"]
fig, axes = plt.subplots(3, 2, figsize=(12, 10))

for i, var in enumerate(variables):
    # Histograma de variables originales
    axes[i, 0].hist(data[var], bins=40, color='skyblue', edgecolor='white')
    axes[i, 0].set_title(f'Distribución Original: {var}')

    # Histograma de variables logarítmicas (usando el nombre con guiones bajos)
    col_log = f"Log_{var.replace(' ', '_')}"
    axes[i, 1].hist(data[col_log], bins=40, color='salmon', edgecolor='white')
    axes[i, 1].set_title(f'Distribución Logarítmica: {col_log}')

plt.tight_layout()
plt.show()

print("\n✅ Análisis completado. Revisa los resultados en la consola y las gráficas.")

print("\n"+"*"*100+"\n")
print("OBJETIVO ESPECIFICO 3: SIMULACIÓN DE ESCENARIO (+10% BASE PRICE)")
print("*"*100+"\n")
```

```python
# 12. OBJETIVO ESPECIFICO 3:
# Modelar un escenario hipotético de la demanda y 
# las unidades vendidas de un incremento del 10% para identificar el comportamiento futuro de las unidades vendidas
```

```python
#13. SE USRA EL MODELO LOG LOG, PORQUE EXPLICA MEJOR EL MODELO
y_real = data['Log_Units_Sold']
X_real = sm.add_constant(data[['Log_Base_Price', 'Log_Total_Price']])
modelo_log_log = sm.OLS(y_real, X_real).fit()

print("\n---  RESUMEN DEL MODELO HISTÓRICO (LOG-LOG) ---")
print("Este es el modelo real sobre el que haremos las predicciones futuras.")
print(modelo_log_log.summary()) # Descomenta esto si quieres ver la tabla completa de nuevo
```

```python
# 14. CREACIÓN DEL ESCENARIO SIMULADO (+10% en Base Price)
data["Base_Price_Sim"] = data["Base Price"] * 1.10
data["Log_Base_Price_Sim"] = np.log(data["Base_Price_Sim"])

# 15. PREDICCIÓN DEL COMPORTAMIENTO FUTURO
# Preparamos las nuevas variables independientes (X) con el precio base alterado
X_simulado = sm.add_constant(data[['Log_Base_Price_Sim', 'Log_Total_Price']])

# 16 USAMOS EL MODELO PREVIAMENTE ENTRENADO PARA PREDECIR SOBRE ESTE NUEVO ESCENARIO 
data['Prediccion_Log_Units'] = modelo_log_log.predict(X_simulado)

# 17. CONVERTIMOS EL RESULTADO LOGARITMICO A UNIDADES FISICAS REALES USANDO EXPONENCIAL
data['Unidades_Futuras_Esperadas'] = np.exp(data['Prediccion_Log_Units'])
```

```python
# 18. RESULTADOS DEL IMPACTO EN LA DEMANDA
unidades_originales = data['Units Sold'].sum()
unidades_futuras = data['Unidades_Futuras_Esperadas'].sum()
diferencia = unidades_futuras - unidades_originales
porcentaje_cambio = (diferencia / unidades_originales) * 100

#19. IMPACTO EN LA DEMANDA
print("\n---  IMPACTO PROYECTADO EN LA DEMANDA ---")
print(f"Unidades vendidas históricas:           {unidades_originales:,.0f}")
print(f"Unidades estimadas con +10% en precio:  {unidades_futuras:,.0f}")
print(f"Impacto total en el volumen de ventas: {diferencia:+,.0f} unidades ({porcentaje_cambio:+.2f}%)")
```

```python
# 20. GRÁFICOS DE COMPARACIÓN
plt.style.use('seaborn-v0_8-muted')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# 21 Gráfico 1: El cambio en el Precio
ax1.hist(data["Base Price"], bins=40, alpha=0.6, label='Precio Base Original', color='#4C72B0')
ax1.hist(data["Base_Price_Sim"], bins=40, alpha=0.6, label='Precio Simulado (+10%)', color='#55A868')
ax1.set_title("1. Simulación: Incremento del Precio Base")
ax1.set_xlabel("Precio Base")
ax1.set_ylabel("Frecuencia")
ax1.legend()

# 22. Gráfico 2: El impacto en las Ventas
# Limitamos el eje X usando el cuantil 95 para evitar que valores extremos aplasten el gráfico
limite_x = data["Units Sold"].quantile(0.95) 
ax2.hist(data[data["Units Sold"] < limite_x]["Units Sold"], bins=40, alpha=0.6, label='Ventas Históricas', color='#C44E52')
ax2.hist(data[data["Unidades_Futuras_Esperadas"] < limite_x]["Unidades_Futuras_Esperadas"], bins=40, alpha=0.6, label='Ventas Proyectadas', color='#8172B3')
ax2.set_title("2. Reacción: Impacto en Unidades Vendidas")
ax2.set_xlabel("Unidades Vendidas")
ax2.legend()

plt.tight_layout()
plt.show()
```
